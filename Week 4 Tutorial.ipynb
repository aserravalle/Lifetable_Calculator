## CFM and UDD Assumptions

```python
#UDD Assumption formulae
def UDD_l(x):
    t = float(x) - int(x) 
    c = (1-t)*l(int(x))+t*l(int(x)+1)
    return c

def UDD_p(x,t=1):
    a = UDD_l(x+t)/UDD_l(x)
    return a

def UDD_q(x, t=1, s=0):
    a = (UDD_l(x+t+s)-UDD_l(x+t+s+1))/UDD_l(x)
    #I'm not certain t and s works (i.e. t|s qx) 
    return a

#CFM Assumption formulae
def CFM_l(x):
    t = float(x) - int(x) 
    c = l(int(x))**(1-t)*l(int(x)+1)**t
    return c

def CFM_p(x,t=1):
    a = CFM_l(x+t)/CFM_l(x)
    return a

def CFM_q(x, t=1, s=0):
    #haven't tested this whole formula
    a = (CFM_l(x+t+s)-CFM_l(x+t+s+1))/CFM_l(x)
    return a
```

## Insurance

```python
def D(x,i):
    v = (1+i)**(-1)
    c = v**x*l(x)
    return c

def WholeLife(x,i):
    v = (1+i)**(-1)
    k = 0
    c = 0
    while k+x in LIFETABLE['x']:
        c += (v**(k+1))*q(x,1,k)
        k += 1
    return c    
    
def PureEndowment(x,n,i):
    c = D(x+n,i)/D(x,i)
    return c

def DefferredLife(x,n,i):
    # not tabulated
    v = (1+i)**(-1)
    c = v**n*p(x,n)*WholeLife(x+n,i)
    return c

def TermInsurance(x,n,i):
    c = WholeLife(x,i) - PureEndowment(x,n,i)*WholeLife(n+x,i)
    return c

def Endowment(x,n,i):
    c = TermInsurance(x,n,i) + PureEndowment(x,n,i)
    return c
```
